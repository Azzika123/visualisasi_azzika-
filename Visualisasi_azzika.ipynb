```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Contoh data
data = {
    "Kelas": ["A"]*10 + ["B"]*10 + ["C"]*10,
    "Nilai": [80,85,78,90,88,75,95,100,82,84,
              70,72,68,75,78,74,80,85,77,73,
              60,65,62,70,68,66,72,74,63,61]
}

df = pd.DataFrame(data)

# --- Boxplot ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)  # subplot 1
sns.boxplot(x="Kelas", y="Nilai", data=df, palette="Set2")
plt.title("Boxplot Nilai per Kelas")

# --- Line Plot ---
plt.subplot(1,2,2)  # subplot 2
sns.lineplot(x=range(1, len(df[df["Kelas"]=="A"])+1),
             y=df[df["Kelas"]=="A"]["Nilai"], marker="o", label="Kelas A")
sns.lineplot(x=range(1, len(df[df["Kelas"]=="B"])+1),
             y=df[df["Kelas"]=="B"]["Nilai"], marker="o", label="Kelas B")
sns.lineplot(x=range(1, len(df[df["Kelas"]=="C"])+1),
             y=df[df["Kelas"]=="C"]["Nilai"], marker="o", label="Kelas C")

plt.title("Line Plot Nilai per Siswa di Setiap Kelas")
plt.xlabel("Siswa ke-")
plt.ylabel("Nilai")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Contoh data
data = {
    "Kelas": ["A"]*10 + ["B"]*10 + ["C"]*10,
    "Nilai": [80,85,78,90,88,75,95,100,82,84,
              70,72,68,75,78,74,80,85,77,73,
              60,65,62,70,68,66,72,74,63,61]
}

df = pd.DataFrame(data)

# Tambah index untuk urutan siswa keseluruhan
df["Urutan"] = range(1, len(df)+1)

# --- Line Plot Tunggal ---
plt.figure(figsize=(8,5))
sns.lineplot(x="Urutan", y="Nilai", data=df, marker="o", color="blue")
plt.title("Line Plot Nilai Semua Kelas")
plt.xlabel("Urutan Siswa (A → B → C)")
plt.ylabel("Nilai")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Contoh data
data = {
    "Kelas": ["A"]*10 + ["B"]*10 + ["C"]*10,
    "Nilai": [80,85,78,90,88,75,95,100,82,84,
              70,72,68,75,78,74,80,85,77,73,
              60,65,62,70,68,66,72,74,63,61]
}

df = pd.DataFrame(data)

# Tambah nomor siswa di tiap kelas (biar sejajar)
df["Siswa ke-"] = df.groupby("Kelas").cumcount() + 1

# --- Lineplot Multiple Line ---
plt.figure(figsize=(8,5))
sns.lineplot(x="Siswa ke-", y="Nilai", hue="Kelas", marker="o", data=df, palette="Set1")
plt.title("Line Plot Multiple Line (Per Kelas)")
plt.xlabel("Siswa ke-")
plt.ylabel("Nilai")
plt.legend(title="Kelas")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import date

# --- Contoh Data Dummy yang BENAR ---
dates = pd.date_range("2021-12-01", "2021-12-31", freq="7D")  # 5 tanggal (1, 8, 15, 22, 29)
n_dates = len(dates)

bank_long = pd.DataFrame({
    "Date": dates.tolist() * 3,  # ulang 3x untuk 3 saham
    "price_rp": [7000, 7200, 7100, 7050, 7150,   # BBRI
                 6500, 6600, 6700, 6650, 6750,   # BMRI
                 7800, 7750, 7700, 7650, 7600],  # BBCA
    "code_stock": ["BBRI"]*n_dates + ["BMRI"]*n_dates + ["BBCA"]*n_dates
})

# --- Visualisasi ---
sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))

# Batas sumbu x dan y
start_date = pd.Timestamp(date(2021, 12, 1))
end_date = pd.Timestamp(date(2021, 12, 31))
plt.xlim(start_date, end_date)
plt.ylim(6000, 8200)

cust_palette = ["#640303", "orange", "c"]

# Line plot
sns.lineplot(
    data=bank_long,
    x="Date", y="price_rp",
    hue="code_stock",
    palette=cust_palette,
    linewidth=3
)

# Marker
sns.scatterplot(
    data=bank_long,
    x="Date", y="price_rp",
    hue="code_stock",
    palette=cust_palette,
    style="code_stock",
    markers={"BBRI": "D", "BMRI": "o", "BBCA": "s"},
    s=120,
    legend=False
)

# Tambah teks di setiap titik
for _, row in bank_long.iterrows():
    plt.text(row["Date"], row["price_rp"]+40, str(row["price_rp"]),
             fontsize=8, ha="center")

plt.title("Pergerakan Harga Saham (Desember 2021)")
plt.xlabel("Tanggal")
plt.ylabel("Harga (Rp)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# --- Pivot data ke bentuk tabel ---
pivot_df = bank_long.pivot(index="Date", columns="code_stock", values="price_rp")

plt.figure(figsize=(8,5))
sns.heatmap(
    pivot_df,
    annot=True, fmt="d",
    cmap="RdPu",      # colormap pink-ungu
    linewidths=0.5,
    cbar_kws={'label': 'Harga Rp'}
)

plt.title("Heatmap Harga Saham (Desember 2021) - Tema Pink")
plt.xlabel("Kode Saham")
plt.ylabel("Tanggal")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- Contoh data dummy ---
dummy_data = pd.DataFrame(
    np.random.randint(60, 100, size=(5, 5)),
    columns=list("ABCDE"),
    index=[f"Row{i}" for i in range(1, 6)]
)

# --- Annotasi teks ---
kws = {
    "fontsize": 9,
    "fontweight": "bold",
    "rotation": 45,
    "ha": "center",
    "va": "center",
}

# --- Heatmap ---
plt.figure(figsize=(6, 5))
sns.heatmap(
    dummy_data,
    annot=True,
    annot_kws=kws,
    linewidths=0.5,   # ✅ perbaikan: harus pakai linewidths
    cmap="RdPu"
)

plt.title("Heatmap dengan Annotasi Custom (Tema Pink)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Contoh data dummy
matrix_data = pd.DataFrame(
    np.random.randint(1, 100, size=(6, 6)),
    columns=list("ABCDEF"),
    index=[f"Row{i}" for i in range(1, 7)]
)

plt.figure(figsize=(7,6))
sns.heatmap(
    matrix_data,
    annot=True, fmt="d",
    cmap=sns.light_palette("pink", as_cmap=True),  # light pink
    linewidths=0.5,
    cbar_kws={'label': 'Nilai'}
)
plt.title("Matriks Plot (Heatmap Matrix - Light Pink)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data dummy
np.random.seed(42)
data = pd.DataFrame({
    "Kategori": np.repeat(["A", "B", "C", "D"], 25),
    "Nilai": np.random.randint(20, 100, 100)
})

plt.figure(figsize=(6,4))
sns.barplot(
    x="Kategori", y="Nilai", data=data,
    palette=sns.light_palette("pink", n_colors=4)
)
plt.title("Barplot - Light Pink")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Kategori", y="Nilai", data=data,
    palette=sns.light_palette("pink", n_colors=4)
)
plt.title("Boxplot - Light Pink")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(
    x="Kategori", y="Nilai", data=data,
    palette=sns.light_palette("pink", n_colors=4)
)
plt.title("Violinplot - Light Pink")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data dummy
np.random.seed(42)
data = pd.DataFrame({
    "Kategori": np.repeat(["A", "B", "C", "D"], 25),
    "Nilai": np.random.randint(20, 100, 100)
})

plt.figure(figsize=(6,4))
sns.stripplot(
    x="Kategori", y="Nilai", data=data,
    palette=sns.light_palette("pink", n_colors=4),
    jitter=True,  # biar titik tidak menumpuk
    size=6        # ukuran titik
)
plt.title("Stripplot - Light Pink")
plt.show()
```
